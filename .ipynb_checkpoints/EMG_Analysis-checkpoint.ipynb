```python
# all df with channel/time data
time_channel_dfs_list = [df_kt, df_mt, df_at, df_atb, df_zt, df_pt, df_yt]
```

```python
# cell compares emg data between participants by running ttest
# params==(df1, df2)
def compute_ttest(df1, df2, channel):
    print("Channel: ", str(channel))
    vals1 = df1.iloc[channel, :].values
    mu1 = np.mean(vals1)
    sigma1 = np.std(vals1)
    vals2 = df2.iloc[channel, :].values
    mu2 = np.mean(vals2)
    sigma2 = np.std(vals2)
    vals1_normalized = normalize(vals1, mu1, sigma1)
    vals2_normalized = normalize(vals2, mu2, sigma2)
    print("DF_PRIMARY\n", mu1, ", ", sigma1, "\nDF_OTHER\n", mu2, ", ", sigma2)
    t, p = ttest_ind(vals1_normalized, vals2_normalized)
    print("p-value: ", p)
    print("t-value: ", t)
```

```python
# cell extracts summary stats from array
# params==(arr)
# returns list of values [mu, std, max, min, var]
def stats_summary(arr):
    mu = np.mean(arr)
    std = np.std(arr)
    mx = np.max(arr)
    mn = np.min(arr)
    var = np.var(arr)
    stats_list = [mu, std, mx, mn, var]
    return stats_list
```

```python
# cell normalizes data for comparisons
def normalize(arr, mu, sigma):
    ret_arr = np.zeros(len(arr))
    for i in range(len(arr)):
        val_normalized = (arr[i]-mu)/sigma
        ret_arr[i] = val_normalized
    return ret_arr
```

```python
def plot_channel(vals, t, title_str):
    plt.title(title_str, FontSize = 20)
    plt.xlabel("Time Slices", FontSize=15)
    plt.ylabel("Signal Value ", FontSize=15)
    plt.plot(t, vals)
    plt.show()
```

```python
# cell visualizes data via plotting values
# params- dataframe and row_index
def visualize(df, row_index):
    time_bins_labels = df.columns
    total_bins = len(time_bins_labels)
    time_arr = get_time_arr(total_bins)
    vals = df.iloc[row_index, :].values # returns n-d vector of values   
    title_str = "Channel " + str(row_index)
    plot_channel(vals, time_arr, title_str)
    stats = stats_summary(vals)
    mu = stats[0]
    std = stats[1]
    mx = stats[2]
    mn = stats[3]
    var = stats[4]
    print(title_str + "\nStats: ")
    print("[Mu:", mu, "Std:", std, "Max:", mx, "Min:", mn, "Var:", var, "]")
    
    
    
    
    
```

```python
# cell creates time array for plotting
def get_time_arr(total_bins):
    time_arr = []
    for i in range(total_bins):
        time_arr.append(i)
    return time_arr
```

```python
# cell controls function execution
def main(data):
    for df in data:
        print("Starting Visualization Process for Dataframe")
        visualize(df, 1)
        visualize(df, 2)
        print("Executing statistical tests for DF...")
        for i in range(len(data)):
            compute_ttest(df, data[i], 1) # stats test channel 1
            compute_ttest(df, data[i], 2) # stats test channel 2
        print('==========================================================\n\n')
```

```python
main(time_channel_dfs_list)
```
